```python
#https://en.wikipedia.org/wiki/Euler%E2%80%93Mascheroni_constant

%matplotlib inline

import matplotlib
import numpy as np

mu = 0
sigma = 1
N = 300
T = 100

cvec = [];

pairs = int(N*(N-1)/2)

for x in range(pairs):
    s1 = np.random.normal(mu, sigma, T);
    s2 = np.random.normal(mu, sigma, T);
    
    s1 = s1 - np.mean(s1);
    s2 = s2 - np.mean(s2);
    v1 = np.dot(s1,s1);
    v2 = np.dot(s2,s2);
    c = np.dot(s1,s2)/(np.sqrt(v1*v2));
    cvec.append(c);


matplotlib.pyplot.hist(cvec)

max(cvec)
```
