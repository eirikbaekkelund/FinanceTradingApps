```python
# https://data.nasdaq.com/tools/python

import nasdaqdatalink
import matplotlib.pyplot as plt

data = nasdaqdatalink.get('NSE/OIL')

data['Open'].plot()
plt.tight_layout()
plt.title('NSE/OIL')
plt.grid()
plt.show()
```
